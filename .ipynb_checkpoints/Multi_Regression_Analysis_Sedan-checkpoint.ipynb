```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

```

```python
# Read in the CSV file as a pandas DataFrame
df = pd.read_csv('Resources/Sedan_model_car_prices.csv')

# Review the DataFrame
df.head()
```

```python
# To drop the columns 'body' and 'mmr' in place
df.drop(['body', 'mmr'], axis=1, inplace=True)
```

```python
# Create a correlation matrix
corr_matrix = df.corr()

# Set the plot size, color, and font scale
sns.set(font_scale=1)
plt.figure(facecolor='gray', figsize=(10, 8))

# Display the matrix without annotations
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm')
plt.show()
```

```python
# Set the plot size and color
plt.figure(facecolor='gray', figsize=(15,8))
ax = plt.axes()
ax.set_facecolor("gray")


# Display a plot of Target correlations
target = corr_matrix['sellingprice']
target.plot(kind='bar')
plt.show()
```

```python
Key = (target > 0.1) | (target < -0.1)
target[Key]
```

```python
# Features to include: 'odometer', 'mmr', 'year', 'make', 'body', 'color', 'interior', 'automatic'

features = ['odometer', 'year', 'make', 'color', 'interior', 'transmission_automatic']

# features = ['odometer', 'year', 'mmr', 'make', 'body', 'color', 'interior', 'automatic']

X = df[features]
X.head()
```

```python
# Set the target variable y
y = df["sellingprice"]
```

```python
# Now split the data into training and testing sets again
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create the model
model = LinearRegression()

# Fit the model to the training data. 
model.fit(X_train, y_train)
```

```python
# Calculate the mean_squared_error and the r-squared value
# for the testing data

# Use our model to make predictions
predicted = model.predict(X_test)

# Score the predictions with mse and r2
mse = mean_squared_error(y_test, predicted)
r2 = r2_score(y_test, predicted)

print(f"mean squared error (MSE): {mse}")
print(f"R-squared (R2): {r2}")
```
