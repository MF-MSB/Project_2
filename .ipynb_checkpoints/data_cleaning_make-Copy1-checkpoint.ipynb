```python
import pandas as pd

# Assuming df is your original DataFrame
# Generate dummy variables for the 'transmission' column
df_transmission = pd.get_dummies(df['transmission'], prefix='transmission')

# Concatenate the original DataFrame with the new dummy variables DataFrame
df = pd.concat([df, df_transmission], axis=1).drop(columns=['transmission_manual'])

# Display the first 5 rows of the updated DataFrame
df.head(5)

```

```python
df.info()
```

```python
# Look for null values
print(df.isnull().sum())
```

```python
# Use Pandas .astype("category").cat.codes for single column category encoding
columns_to_encode = ["body",
                     "color",
                     "interior"]

# Loop through columns_to_encode and convert the columns to category codes
for column in columns_to_encode:
    df[column] = df[column].astype("category").cat.codes

df.head()
```

```python
df.info()
```

```python
# Group by the 'make' column
grouped = df.groupby('make')

# Initialize an empty dictionary to store the dataframes
dfs = {}

# Loop through the groups
for make, group_df in grouped:
    # Assign the group dataframe to the corresponding make key
    dfs[make] = group_df
```

```python
# # Assuming `dfs` is your dictionary of DataFrames, grouped by 'make'
# columns = ['odometer', 'mmr']  # Columns to clean and scale

# for make, df in dfs.items():
#     # Step 1: Remove outliers
#     z_scores = np.abs(stats.zscore(df[columns], nan_policy='omit'))
#     threshold = 3
#     without_outliers = (z_scores < threshold).all(axis=1)
#     df_clean = df[without_outliers]

#     # Step 2: Scale the 'odometer' and 'mmr' columns
#     scaler = StandardScaler()
#     df_clean_scaled = df_clean.copy()
#     df_clean_scaled[columns] = scaler.fit_transform(df_clean[columns])

#     # Step 3: Write the cleaned and scaled DataFrame to a CSV file
#     # Customize the file path and name as needed
#     file_name = f'Resources/{make}_model_car_prices.csv'
#     df.to_csv(file_name, index=False)

#     print(f'Processed and saved: {file_name}')


# # Assuming `dfs` is your dictionary of DataFrames, grouped by 'make'
# columns = ['odometer', 'mmr']  # Columns to clean and scale

# for make, df in dfs.items():
#     print(f"Processing make: {make}")

#     # Remove outliers
#     z_scores = np.abs(stats.zscore(df[columns], nan_policy='omit'))
#     threshold = 3
#     without_outliers = (z_scores < threshold).all(axis=1)
#     df_clean = df[without_outliers]

#     # Scale the 'odometer' and 'mmr' columns
#     scaler = StandardScaler()
#     df_clean_scaled = df_clean.copy()
#     df_clean_scaled[columns] = scaler.fit_transform(df_clean[columns])

#     # Output the .info() of the cleaned and scaled DataFrame
#     print(f"Info for {make}:")
#     df_clean_scaled.info()

#     # Optionally, write the DataFrame to a CSV file
#     # Customize the file path and name as needed
#     file_name = f'Resources/{make}_model_car_prices.csv'
#     df_clean_scaled.to_csv(file_name, index=False)

#     print(f'Processed and saved: {file_name}\n\n')

from scipy import stats
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
import os

# Assuming `dfs` is your dictionary of DataFrames, grouped by 'make'
columns = ['odometer', 'mmr']  # Columns to clean and scale

# Initialize a dictionary to hold the make and file size
file_sizes = {}

for make, df in dfs.items():
    print(f"Processing make: {make}")

    # Remove outliers
    z_scores = np.abs(stats.zscore(df[columns], nan_policy='omit'))
    threshold = 3
    without_outliers = (z_scores < threshold).all(axis=1)
    df_clean = df[without_outliers]

    # Scale the 'odometer' and 'mmr' columns
    scaler = StandardScaler()
    df_clean_scaled = df_clean.copy()
    df_clean_scaled[columns] = scaler.fit_transform(df_clean[columns])

    # Output the .info() of the cleaned and scaled DataFrame
    print(f"Info for {make}:")
    df_clean_scaled.info()

    # Write the DataFrame to a CSV file
    # Customize the file path and name as needed
    file_name = f'Resources/{make}_model_car_prices.csv'
    df_clean_scaled.to_csv(file_name, index=False)

    # Get the file size in bytes and convert it to kilobytes (1 KB = 1024 bytes)
    file_size_kb = os.path.getsize(file_name) / 1024
    file_sizes[make] = file_size_kb

    print(f'Processed and saved: {file_name}\n\n')

# After processing all makes, write the file sizes to a new CSV file
sizes_df = pd.DataFrame(list(file_sizes.items()), columns=['Make', 'FileSizeKB'])
sizes_file_name = 'Resources/make_file_sizes.csv'
sizes_df.to_csv(sizes_file_name, index=False)

print(f'File sizes written to: {sizes_file_name}')















```
