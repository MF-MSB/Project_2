```python
import re

# Define the function to extract the year
def extract_year(sale_date):
    pattern = r'\b\d{4}\b'
    match = re.search(pattern, sale_date)
    if match:
        return int(match.group())  # Convert the matched year to an integer
    else:
        return None  # Return None if no year is found

# Apply the function to the 'saledate' column to create a new 'saleyear' column in the existing DataFrame
df['saleyear'] = df['saledate'].apply(extract_year)

# df now has a new column 'saleyear' in addition to its original columns

df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame and it's already loaded with 'year' and 'make' columns

# Group by 'year' and 'make', then count the occurrences
data_per_make_per_year = df.groupby(['year', 'make']).size().reset_index(name='counts')

# Pivot the data to have years as rows, makes as columns, and counts as values
pivot_data = data_per_make_per_year.pivot(index='year', columns='make', values='counts').fillna(0)

# Plotting
plt.figure(figsize=(10, 6)) # Adjust the size as needed
for make in pivot_data.columns:
    plt.plot(pivot_data.index, pivot_data[make], marker='', label=make)

plt.title('Number of Occurrences per Make Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Occurrences')
plt.legend(title='Make', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45) # Rotate x-axis labels for better readability
plt.tight_layout() # Adjust layout to make room for the legend
plt.show()
```
