```python
import re

# Define the function to extract the year
def extract_year(sale_date):
    pattern = r'\b\d{4}\b'
    match = re.search(pattern, sale_date)
    if match:
        return int(match.group())  # Convert the matched year to an integer
    else:
        return None  # Return None if no year is found

# Apply the function to the 'saledate' column to create a new 'saleyear' column in the existing DataFrame
df['saleyear'] = df['saledate'].apply(extract_year)

# df now has a new column 'saleyear' in addition to its original columns

df.head()
```

```python
# Assuming df is your DataFrame and it includes the 'make' and 'body' columns

# Calculate the count of each body type within each make directly
df_counted = df.groupby(['make', 'body']).size().reset_index(name='count')

# Calculate the percentage of each body type within each make
df_counted['percentage'] = df_counted.groupby('make')['count'].transform(lambda x: x / x.sum() * 100)

# Pivot the table to have makes as indexes and body types as columns, with percentages as values
pivot_df = df_counted.pivot(index='make', columns='body', values='percentage').fillna(0)

# Calculate total counts for each make to determine the order
total_counts = df_counted.groupby('make')['count'].sum()

# Sort the pivot_df according to the total counts, descending
pivot_df = pivot_df.reindex(total_counts.sort_values(ascending=False).index)

# Plotting
pivot_df.plot(kind='bar', stacked=True, figsize=(10, 7), colormap='winter')
plt.title('Normalized Stacked Bar Chart of Car Bodies by Make')
plt.xlabel('Car Make')
plt.ylabel('Percentage')
plt.xticks(rotation=90)
plt.gca().yaxis.set_major_formatter(plt.matplotlib.ticker.PercentFormatter())

# Move the legend to the right-hand side of the chart
plt.legend(title='Body Type', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout to make room for the legend
plt.tight_layout()



# Saving the adjusted plot to a file BEFORE showing it
plt.savefig('Normalized_Stacked_Bar_Chart_of_Car+Bodies_by_Make.png')

# Now show the plot
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming df is your DataFrame and it includes the 'make', 'sellingprice' columns

# Step 1: Sum the selling prices for each 'make'
summed_prices = df.groupby('make')['sellingprice'].sum()

# Step 2: Count the number of entries for each 'make'
entry_counts = df['make'].value_counts()

# Ensure summed_prices is sorted by the same order as entry_counts (descending frequency)
# We sort entry_counts in descending order and then reindex summed_prices based on this
summed_prices_sorted = summed_prices.reindex(entry_counts.index)

# Plotting the vertical bar chart
plt.figure(figsize=(10, 7))
summed_prices_sorted.plot(kind='bar')
plt.title('Summed Selling Prices by Car Make')
plt.ylabel('Summed Selling Price')
plt.xlabel('Car Make')
plt.xticks(rotation=90)  # Rotate the x-axis labels for better readability

# Saving the adjusted plot to a file BEFORE showing it
plt.savefig('Summed_Selling_Prices_by_Car_Make.png')

plt.show()


```
